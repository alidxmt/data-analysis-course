#### List - Loop
Listen speichern mehrere Elemente. Wir können sie mit einer for-loop durchlaufen.

```python
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit.upper())
```

#### Range-Loop

```python
for i in range(5):
    if i == 3:
        continue
    print(i)
```

#### Bedingte Anweisungen 
Bedingte Anweisungen (Conditional statements) erlauben es uns, Code basierend auf einer Bedingung auszuführen

```python
x = 10
if x > 0:
    print("Positive")
elif x == 0:
    print("Zero")
else:
    print("Negative")
```

#### Funktion 
Funktionen erlauben es uns, Code in wiederverwendbare Blöcke zu organisieren. Wir können Parameter übergeben und Werte zurückgeben.

```python
def greet(name="World"):
    return f"Hello, {name}!"

print(greet("Eva"))
```

#### Analyse von Wortlisten mit Kommentaren (Funktion)

Diese Funktion durchläuft eine Liste von Wörtern.
Sie gibt aus, ob ein Wort kurz, mittel oder lang ist und markiert Wörter, die den Buchstaben 'a' enthalten.

Aufgabe: Schreibe vor jedem Codeblock einen Kommentar, der in einfachen Worten erklärt,
was der nächste Codeblock macht. 

```python
def analyze_words(words):
    for word in words:
        # Check word length, len is a function that return length of a string
        # ----------------------------here explain what happens in if-elseif-if block
        if len(word) <= 3:
            size = "short"
        elif len(word) <= 6:
            size = "medium"
        else:
            size = "long"
        
        # Check if word contains 'a' with (x in word.lower()). 
        # We turn all to lower case to check whther it has 'a',
        # Otherwise we would miss words like "Apple" as 'a' is uppercase there.
        if 'a' in word.lower():
            print(f"{word.upper()} ({size}) contains 'a'! 🌟")
        else:
            print(f"{word} ({size})")

# Example usage of the function
my_words = ["cat", "Python", "AI", "banana", "sun", "fantastic"]

analyze_words(my_words)
```
