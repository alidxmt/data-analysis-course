```python
# Food Composition / Nutrient Database

Overview  
This dataset contains detailed nutrient composition for individual foods. Each row is one food item (Category + Description) identified by a Nutrient Data Bank Number. Values give measured macronutrients, micronutrients, pigments, and serving/household weight information.

Key columns (what they represent)
- Category, Description, Nutrient Data Bank Number — food identity and ID.  
- Energy & macronutrients: Data.Kilocalories, Data.Protein, Data.Carbohydrate, Data.Fat.Total Lipid, Data.Fiber, Data.Sugar Total, Data.Water, Data.Cholesterol, Data.Choline.  
- Fat profile: Data.Fat.Monosaturated Fat, Data.Fat.Polysaturated Fat, Data.Fat.Saturated Fat.  
- Fat‑soluble and pigment compounds: Data.Alpha Carotene, Data.Beta Carotene, Data.Beta Cryptoxanthin, Data.Lutein and Zeaxanthin, Data.Lycopene, Data.Retinol.  
- Vitamins: Data.Vitamins.Vitamin A - IU, Data.Vitamins.Vitamin A - RAE, Data.Vitamins.Vitamin B12, Data.Vitamins.Vitamin B6, Data.Vitamins.Vitamin C, Data.Vitamins.Vitamin E, Data.Vitamins.Vitamin K, plus Niacin, Riboflavin, Thiamin, Pantothenic Acid.  
- Major minerals and trace elements: Data.Major Minerals.Calcium, Copper, Iron, Magnesium, Phosphorus, Potassium, Sodium, Zinc, plus Data.Manganese and Data.Selenium.  
- Practical metadata: Data.Refuse Percentage (edible fraction), Data.Household Weights.* (common serving sizes and descriptions).

What it's good for
- Calculating nutrient totals per serving or per 100 g (single foods or aggregated recipes).  
- Comparing nutrient density across foods or categories (e.g., highest iron, vitamin A).  
- Diet planning, menu analysis, and nutrition labeling.  
- Public‑health or epidemiological analyses of food supply/nutrient intake.  
- Feature data for ML tasks: food clustering, nutrient‑based recommendations, predicting missing nutrient values.

Typical workflows
- Filter by Category/Description, compute per‑serving nutrients using Household Weights.  
- Normalize to per‑100 g using Water/Refuse percentage and weight columns.  
- Rank or visualize foods by selected nutrients or nutrient ratios.  
- Aggregate by category for summary statistics.
```
