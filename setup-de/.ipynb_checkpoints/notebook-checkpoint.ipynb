# Data

#### Definition von Daten in der Informatik

Beginnen wir mit der Definition von **Daten** im Wörterbuch der Informatik ([foldoc.org/data](http://foldoc.org/data)):

> Daten sind „Zahlen, Zeichen, Bilder oder andere Aufzeichnungsformen, in einer Form, die von Menschen interpretiert oder – insbesondere – in einen Computer eingegeben, dort gespeichert, verarbeitet oder über digitale Kanäle übertragen werden kann. Computer stellen Daten nahezu immer in Binärform dar.  
> Daten für sich haben keine Bedeutung; erst durch die Interpretation in einem Datenverarbeitungssystem erhalten sie Bedeutung und werden zu Information.“



---

#### Daten vs. Information

Im alltäglichen Sprachgebrauch werden **Daten** und **Informationen** häufig synonym verwendet, obwohl sie unterschiedliche Konzepte darstellen:  

- **Daten** werden erst durch Interpretation in einem Verarbeitungssystem zu **Information**.  

Beispiel: Suchen Sie den Begriff **„Pinguin“** (oder ein Metall wie „Eisen“) in:

- [Google Dataset Search](https://datasetsearch.research.google.com/)  
- [Google-Suche](https://www.google.com/)  

Welche Suche liefert Ihnen tatsächlich relevante Informationen?

#### Etymologie des Begriffs „Daten“

- Aus dem Lateinischen: **datum** – „gegebenes Ding“  
- **dare (Latein)**: geben  
- **datum**: gegebenes Ding  
- **datum (1640er Jahre)**: eine gegebene Tatsache  
  - Quelle: [etymonline.com](https://www.etymonline.com)
