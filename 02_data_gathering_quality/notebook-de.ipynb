In diesem Notebook lernen wir, wie man mit Python Daten von APIs abruft.

### Der API-Zyklus
Wenn wir eine API anfragen, senden wir einen **Request** (Anfrage) und erhalten einen **Response** (Antwort) zurück, meist im JSON-Format.



## Vorbereitung
Wir nutzen die `requests` Bibliothek, um die Daten zu laden.

```python
import requests
from IPython.display import Image, display

print("Setup abgeschlossen. Bibliotheken sind geladen.")
```

## Beispiel 1: JSONPlaceholder (Kostenlose Test-Daten)
Ideal, um zu prüfen, ob die Internetverbindung steht.

```python
url = 'https://jsonplaceholder.typicode.com/posts/1'
response = requests.get(url)

if response.status_code == 200:
    data = response.json()
    print(f"Titel des Posts: {data['title']}")
else:
    print(f"Fehler: {response.status_code}")
```

## Beispiel 2: Open-Meteo (Wetter - Ohne Key)
**Hinweis:** Dies ist der Ersatz für die alte MetaWeather API.

```python
# Wetter für Berlin abrufen (Breitengrad: 52.52, Längengrad: 13.41)
weather_url = "https://api.open-meteo.com/v1/forecast?latitude=52.52&longitude=13.41&current_weather=true"
response = requests.get(weather_url)

if response.status_code == 200:
    data = response.json()
    temp = data['current_weather']['temperature']
    wind = data['current_weather']['windspeed']
    print(f"Aktuelle Temperatur in Berlin: {temp}°C")
    print(f"Windgeschwindigkeit: {wind} km/h")
else:
    print("Wetterdaten konnten nicht geladen werden.")
```

## Beispiel 3: Dog CEO (Bilder verarbeiten)
Hier lernen wir, wie man eine Bild-URL aus einer API-Antwort extrahiert und anzeigt.

```python
url = 'https://dog.ceo/api/breeds/image/random'
data = requests.get(url).json()
img_url = data['message']

print(f"Bild-URL: {img_url}")
display(Image(url=img_url, width=300))
```

## Beispiel 4: CoinGecko (Kryptowährungen)
Echtzeit-Preise für Bitcoin in USD.

```python
crypto_url = 'https://api.coingecko.com/api/v3/simple/price?ids=bitcoin&vs_currencies=usd'
response = requests.get(crypto_url)

if response.status_code == 200:
    preis = response.json()['bitcoin']['usd']
    print(f"Aktueller Bitcoin Preis: ${preis:,.2f}")
else:
    print("Anfragelimit erreicht? Bitte eine Minute warten.")
```

### Tipps für die Praxis:
- **Status Codes:** Ein Code `200` bedeutet Erfolg, `404` heißt 'Nicht gefunden' und `401` bedeutet 'Nicht autorisiert' (Key fehlt).
- **JSON:** Die meisten APIs antworten in einem Format, das wie ein Python-Dictionary aussieht.
