# Learning APIs with Python 🐍 (Updated 2026)

This notebook demonstrates how to interact with modern REST APIs. 

### API Communication Overview
When you use a library like `requests`, you are participating in a **Request-Response Cycle**.



```python
import requests
from IPython.display import Image, display

print("Setup complete. Libraries imported.")
```

## Example 1: JSONPlaceholder (Mock Data)
Great for testing if your connection works.

```python
url = 'https://jsonplaceholder.typicode.com/posts/1'
response = requests.get(url)

if response.status_code == 200:
    data = response.json()
    print(f"Post Title: {data['title']}")
else:
    print(f"Error: {response.status_code}")
```

## Example 2: Open-Meteo (Weather - No Key Required)
**Note:** This replaces the retired MetaWeather API.

```python
# Fetching weather for Berlin (Lat: 52.52, Long: 13.41)
weather_url = "https://api.open-meteo.com/v1/forecast?latitude=52.52&longitude=13.41&current_weather=true"
response = requests.get(weather_url)

if response.status_code == 200:
    data = response.json()
    temp = data['current_weather']['temperature']
    print(f"Current Temperature in Berlin: {temp}°C")
else:
    print("Could not fetch weather data.")
```

## Example 3: Dog CEO (Images)
How to handle image URLs from an API.

```python
url = 'https://dog.ceo/api/breeds/image/random'
data = requests.get(url).json()
img_url = data['message']

print(f"Showing dog from: {img_url}")
display(Image(url=img_url, width=300))
```

## Example 4: CoinGecko (Crypto Prices)
Uses a public API to get real-time Bitcoin prices.

```python
crypto_url = 'https://api.coingecko.com/api/v3/simple/price?ids=bitcoin&vs_currencies=usd'
response = requests.get(crypto_url)

if response.status_code == 200:
    price = response.json()['bitcoin']['usd']
    print(f"Bitcoin Current Price: ${price:,.23f}")
else:
    print("Rate limit hit? Try again in a minute.")
```
