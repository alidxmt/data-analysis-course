# Data

## Definition of Data

Let's start with the definition of **data** according to the dictionary of computer science ([foldoc.org/data](http://foldoc.org/data)):

> Data are "numbers, characters, images, or other recorded forms, in a form that can be interpreted by humans or—especially—input into a computer, stored, processed, or transmitted over digital channels. Computers almost always represent data in binary form.  
> Data by themselves have no meaning; only through interpretation in a data processing system do they acquire meaning and become information."

## Data Types
Data types define what kind of data a variable can store and how the computer processes it.  
We also explain best practices for naming variables.

### **Integer**: whole numbers  
  - **Naming:** descriptive noun, lowercase, **snake_case**  
  - **Example:** `age = 25`, `score = 100`  
  - **Tip:** avoid single-letter names like `x` unless in loops

```python
# Integer
# Naming: descriptive noun, lowercase, snake_case
# Task: create your own integer variable and print it
age = 25
print("age =", age, "| Type:", type(age))
```

### **Float**: decimal numbers  
  - **Naming:** descriptive noun, lowercase, **snake_case**  
  - **Example:** `temperature = 36.6`, `price = 9.99`  
  - **Tip:** include units in the variable name if relevant (`temperature_celsius`)

```python
# Float
# Naming: descriptive noun, lowercase, snake_case
# Task: create your own float variables and print them
temperature = 36.6
price = 9.99
print("temperature =", temperature, "| Type:", type(temperature))
print("price =", price, "| Type:", type(price))
```

### **String**: text  
  - **Naming:** descriptive noun, lowercase, **snake_case**  
  - **Example:** `username = "Ali"`, `city_name = "Berlin"`  
  - **Tip:** use names that clearly indicate the content

```python
# String
# Naming: descriptive noun, lowercase, snake_case
# Task: create your own string variables and print them
username = "Thomas"
city_name = "Berlin"
print("username =", username, "| Type:", type(username))
print("city_name =", city_name, "| Type:", type(city_name))
```

### **Boolean**: true/false  
  - **Naming:** start with prefix like `is_`, `has_`, `can_`, `should_`, lowercase, **snake_case**  
  - **Example:** `is_active = True`, `has_permission = False`, `can_edit = True`  
  - **Tip:** prefix makes meaning self-explanatory

```python
# Boolean
# Naming: prefix is_/has_/can_/should_, lowercase, snake_case
# Task: create your own boolean variables and print them
is_active = True
has_permission = False
can_edit = True
print("is_active =", is_active, "| Type:", type(is_active))
print("has_permission =", has_permission, "| Type:", type(has_permission))
print("can_edit =", can_edit, "| Type:", type(can_edit))
```

### **List / Array**: ordered collection  
  - **Naming:** plural noun, lowercase, **snake_case**; optional `_list` suffix  
  - **Example:** `users = ["Ali", "Sara"]`, `scores = [10, 20, 30]`  
  - **Tip:** plural signals collection; keep naming consistent

```python
# List/Array
# Naming: plural noun, lowercase, snake_case, optional _list suffix
# Task: create your own list and print it
users = ["Ali", "Sara"]
scores = [10, 20, 30]
print("users =", users, "| Type:", type(users))
print("scores =", scores, "| Type:", type(scores))
```

### **Tuple**: immutable ordered collection  
  - **Naming:** plural noun, lowercase, **snake_case**; optional `_tuple` suffix  
  - **Example:** `coordinates = (10, 20)`, `rgb_colors = (255, 0, 0)`  
  - **Tip:** emphasize immutability with `_tuple` suffix if needed

```python
# Tuple
# Naming: plural noun, lowercase, snake_case, optional _tuple suffix
# Task: create your own tuple and print it
coordinates = (10, 20)
rgb_colors = (255, 0, 0)
print("coordinates =", coordinates, "| Type:", type(coordinates))
print("rgb_colors =", rgb_colors, "| Type:", type(rgb_colors))
```

### **Dictionary**: key-value pairs  
  - **Naming:** descriptive noun + `_dict` suffix, lowercase, **snake_case**  
  - **Example:** `user_dict = {"name":"Ali", "age":30}`, `config_dict = {"theme":"dark"}`  
  - **Tip:** suffix clarifies type; keys should be meaningful

```python
# Dictionary
# Naming: descriptive noun + _dict suffix, lowercase, snake_case
# Task: create your own dictionary and print it
user_dict = {"name":"Ali", "age":30}
config_dict = {"theme":"dark"}
print("user_dict =", user_dict, "| Type:", type(user_dict))
print("config_dict =", config_dict, "| Type:", type(config_dict))
```

### **Set**: unique values  
  - **Naming:** descriptive noun + `_set` suffix, lowercase, **snake_case**  
  - **Example:** `unique_ids = {1, 2, 3}`, `color_set = {"red", "blue"}`  
  - **Tip:** suffix signals type; improves readability

```python
# Set
# Naming: descriptive noun + _set suffix, lowercase, snake_case
# Task: create your own set and print it
unique_ids = {1, 2, 3}
color_set = {"red", "blue"}
print("unique_ids =", unique_ids, "| Type:", type(unique_ids))
print("color_set =", color_set, "| Type:", type(color_set))
```

### **None / Null**: no value  
  - **Naming:** descriptive noun, lowercase, **snake_case**  
  - **Example:** `selected_item = None`, `current_user = None`  
  - **Tip:** use meaningful names to indicate optional or unset values

```python
# None / Null
# Naming: descriptive noun, lowercase, snake_case
# Task: create your own None variables and print them
selected_item = None
current_user = None
print("selected_item =", selected_item, "| Type:", type(selected_item))
print("current_user =", current_user, "| Type:", type(current_user))
```

## Categories of Data

- **Primary data** – collected directly from the source, e.g., survey responses, measurements  
- **Derived data** – generated or summarized from other data, e.g., statistics, reports, aggregated datasets  
- **Operational data** – used to support ongoing processes, e.g., transaction data  
- **Metadata** – data about data, describing structure, origin, or attributes, e.g., creation date, file size

## Semantic Information
**R is information if:**
  - R contains at least one datum
  - R is well-formed
  - R has meaning

Factal vs. instructional information
- **Factual information** – can be **true** or **false**  
- **Instructional information** – can be **useful** or **trustworthy**

## Categorical (Nominal / Ordinal)

**Explanation:**  
- **Nominal:** categories **without order**  
  - Example: snack names, house colors (`Red`, `Blue`)  
- **Ordinal:** categories with a **natural order**  
  - Example: spice levels (`Mild` < `Medium` < `Hot` < `Extreme`)  

**Difference:**  
- Nominal = labels only, **no ranking**  
- Ordinal = labels with **meaningful order**

| Column       | Type   | Category / Values                             | Example                                  |
|--------------|--------|-----------------------------------------------|------------------------------------------|
| spiciness    | string | Ordinal: "Mild" < "Medium" < "Hot" < "Extreme" | "Hot"                                    |
| snack_type   | string | Nominal: no order, just labels               | "Chips", "Chocolate", "Candy"           |
