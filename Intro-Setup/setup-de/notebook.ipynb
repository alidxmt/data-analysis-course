# Data

## Definition von Daten

Beginnen wir mit der Definition von **Daten** im Wörterbuch der Informatik ([foldoc.org/data](http://foldoc.org/data)):

> Daten sind „Zahlen, Zeichen, Bilder oder andere Aufzeichnungsformen, in einer Form, die von Menschen interpretiert oder – insbesondere – in einen Computer eingegeben, dort gespeichert, verarbeitet oder über digitale Kanäle übertragen werden kann. Computer stellen Daten nahezu immer in Binärform dar.  
> Daten für sich haben keine Bedeutung; erst durch die Interpretation in einem Datenverarbeitungssystem erhalten sie Bedeutung und werden zu Information.“



## Datentypen
Datentypen legen fest, welche Art von Daten eine Variable speichern kann und wie der Computer sie verarbeitet.  
Hier erklären wir auch, wie Variablen sinnvoll benannt werden sollten.



### **Integer (Ganzzahl)**: ganze Zahlen  
  - **Naming:** descriptive noun, lowercase, **snake_case**  
  - **Example:** `age = 25`, `score = 100`  
  - **Tip:** avoids vague letters like `x` unless in loops

```python
# Integer (Ganzzahl)
# Benennung: beschreibendes Nomen, Kleinbuchstaben, snake_case
# Aufgabe: erstelle deine eigene Integer-Variable und gib sie aus
age = 25
print("age =", age, "| Typ:", type(age))
```


### **Float (Gleitkommazahl)**: Dezimalzahlen  
  - **Naming:** descriptive noun, lowercase, **snake_case**  
  - **Example:** `temperature = 36.6`, `price = 9.99`  
  - **Tip:** often paired with units in name if relevant (`temperature_celsius`)


```python
# Float (Gleitkommazahl)
# Benennung: beschreibendes Nomen, Kleinbuchstaben, snake_case
# Aufgabe: erstelle deine eigene Float-Variable und gib sie aus
temperature = 36.6
price = 9.99
print("temperature =", temperature, "| Typ:", type(temperature))
print("price =", price, "| Typ:", type(price))
```


### **String (Zeichenkette)**: Text  
  - **Naming:** descriptive noun, lowercase, **snake_case**  
  - **Example:** `username = "Ali"`, `city_name = "Berlin"`  
  - **Tip:** use names that clearly indicate the content

```python
# String (Zeichenkette)
# Benennung: beschreibendes Nomen, Kleinbuchstaben, snake_case
# Aufgabe: erstelle deine eigene String-Variable und gib sie aus
username = "Thomas"
city_name = "Berlin"
print("username =", username, "| Typ:", type(username))
print("city_name =", city_name, "| Typ:", type(city_name))
```


### **Boolean (Boolesch)**: wahr/falsch  
  - **Naming:** start with **prefix** like `is_`, `has_`, `can_`, `should_`, lowercase, **snake_case**  
  - **Example:** `is_active = True`, `has_permission = False`, `can_edit = True`  
  - **Tip:** prefix makes meaning self-explanatory


```python
# Boolean (Boolesch)
# Benennung: Präfix is_/has_/can_/should_, Kleinbuchstaben, snake_case
# Aufgabe: erstelle deine eigene Boolean-Variable und gib sie aus
is_active = True
has_permission = False
can_edit = True
print("is_active =", is_active, "| Typ:", type(is_active))
print("has_permission =", has_permission, "| Typ:", type(has_permission))
print("can_edit =", can_edit, "| Typ:", type(can_edit))
```

### **List/Array (Liste/Array)**: geordnete Sammlung  
  - **Naming:** plural noun, lowercase, **snake_case**; optional `_list` suffix  
  - **Example:** `users = ["Ali", "Sara"]`, `scores = [10, 20, 30]`  
  - **Tip:** plural signals collection; keep consistent naming


```python
# List/Array (Liste/Array)
# Benennung: Plural-Nomen, Kleinbuchstaben, snake_case, optional _list-Suffix
# Aufgabe: erstelle deine eigene Liste und gib sie aus
users = ["Ali", "Sara"]
scores = [10, 20, 30]
print("users =", users, "| Typ:", type(users))
print("scores =", scores, "| Typ:", type(scores))
```


### **Tuple (Tupel)**: unveränderliche geordnete Sammlung  
  - **Naming:** plural noun, lowercase, **snake_case**; optional `_tuple` suffix  
  - **Example:** `coordinates = (10, 20)`, `rgb_colors = (255, 0, 0)`  
  - **Tip:** emphasize immutability with `_tuple` suffix if needed


```python
# Tuple (Tupel)
# Benennung: Plural-Nomen, Kleinbuchstaben, snake_case, optional _tuple-Suffix
# Aufgabe: erstelle dein eigenes Tuple und gib es aus
coordinates = (10, 20)
rgb_colors = (255, 0, 0)
print("coordinates =", coordinates, "| Typ:", type(coordinates))
print("rgb_colors =", rgb_colors, "| Typ:", type(rgb_colors))
```


### **Dictionary (Wörterbuch)**: Key-Value-Paare  
  - **Naming:** descriptive noun + `_dict` suffix, lowercase, **snake_case**  
  - **Example:** `user_dict = {"name":"Ali", "age":30}`, `config_dict = {"theme":"dark"}`  
  - **Tip:** suffix clarifies type; keys should be meaningful


```python
# Dictionary (Wörterbuch)
# Benennung: beschreibendes Nomen + _dict-Suffix, Kleinbuchstaben, snake_case
# Aufgabe: erstelle dein eigenes Dictionary und gib es aus
user_dict = {"name":"Ali", "age":30}
config_dict = {"theme":"dark"}
print("user_dict =", user_dict, "| Typ:", type(user_dict))
print("config_dict =", config_dict, "| Typ:", type(config_dict))
```


### **Set (Menge)**: eindeutige Werte  
  - **Naming:** descriptive noun + `_set` suffix, lowercase, **snake_case**  
  - **Example:** `unique_ids = {1, 2, 3}`, `color_set = {"red", "blue"}`  
  - **Tip:** suffix signals type; useful for readability


```python
# Set (Menge)
# Benennung: beschreibendes Nomen + _set-Suffix, Kleinbuchstaben, snake_case
# Aufgabe: erstelle dein eigenes Set und gib es aus
unique_ids = {1, 2, 3}
color_set = {"red", "blue"}
print("unique_ids =", unique_ids, "| Typ:", type(unique_ids))
print("color_set =", color_set, "| Typ:", type(color_set))
```


### **None/Null (Nullwert)**: kein Wert  
  - **Naming:** descriptive noun, lowercase, **snake_case**  
  - **Example:** `selected_item = None`, `current_user = None`  
  - **Tip:** use meaningful name to indicate optional or unset value

```python
# None/Null (Nullwert)
# Benennung: beschreibendes Nomen, Kleinbuchstaben, snake_case
# Aufgabe: erstelle deine eigene None-Variable und gib sie aus
selected_item = None
current_user = None
print("selected_item =", selected_item, "| Typ:", type(selected_item))
print("current_user =", current_user, "| Typ:", type(current_user))
```

## Kategorien von Daten

- **Primärdaten** – Daten, die direkt aus der Quelle erhoben werden; z. B. Umfrageantworten, Messwerte  
- **Derivative Daten** – Daten, die aus anderen Daten abgeleitet oder zusammengefasst werden; z. B. Statistiken, Berichte, aggregierte Daten  
- **Operationale Daten** – Daten, die zur Unterstützung laufender Prozesse und Abläufe genutzt werden; z. B. Transaktionsdaten  
- **Metadaten** – Daten über Daten, die deren Struktur, Herkunft oder Eigenschaften beschreiben; z. B. Erstellungsdatum, Dateigröße

## Semantic Information
**R ist Information, wenn:**
  - R mindestens ein Datum enthält
  - R wohlgeformt ist
  - R Bedeutung hat

Factal vs. instructional information
- **Faktische Information** – Kann **wahr** oder **falsch** sein  
- **Instruktionale Information** – Kann **hilfreich** oder **vertrauenswürdig** sein

## Kategorisch (Nominal / Ordinal)

**Erklärung:**  
- **Nominal:** Kategorien **ohne Reihenfolge**.  
  - Beispiel: Snack-Namen, Hausfarben (`Rot`, `Blau`).  
- **Ordinal:** Kategorien mit **natürlicher Reihenfolge**.  
  - Beispiel: Schärfegrade (`Mild` < `Medium` < `Hot` < `Extreme`).  

**Unterschied:**  
- Nominal = nur Labels, **keine Rangfolge**.  
- Ordinal = Labels mit **bedeutungsvoller Reihenfolge**.

| Column       | Type   | Category / Values                             | Example                                  |
|--------------|--------|-----------------------------------------------|------------------------------------------|
| spiciness    | string | Ordinal: "Mild" < "Medium" < "Hot" < "Extreme" | "Hot"                                    |
| snack_type   | string | Nominal: no order, just labels               | "Chips", "Chocolate", "Candy"           |

```python
# hier ist ein Dictionary (ein Wörterbuch) spiciness ist ordinal und snack_type ist nominal.

Cheesy_Unicorn_Puffs_snack = {
    "snack_name": "Cheesy Unicorn Puffs",
    "snack_type": "Chips",
    "spiciness": "Medium",
    "price_usd": 4.75,
    "is_vegan": False,
    "taste_description": "Tastes like rainbow dreams with a hint of regret"
}

# Zugriff auf die Eigenschaft "spiciness"
Cheesy_Unicorn_Puffs_snack["spiciness"]
```

Oben war ein Dictionary (ein Wörterbuch), und folgendes sind zwei Snacks mit verschiedenen Eigenschaften.
Wir können diese Daten in eine Tabelle umwandeln.

```python
snacks = [
    {
        "snack_name": "Cheesy Unicorn Puffs",
        "snack_type": "Chips",
        "spiciness": "Medium",
        "price_usd": 4.75,
        "is_vegan": False,
        "taste_description": "Tastes like rainbow dreams with a hint of regret"
    },
    {
        "snack_name": "Soggy Spaghetti Chips",
        "snack_type": "Chips",
        "spiciness": "Hot",
        "price_usd": 3.50,
        "is_vegan": True,
        "taste_description": "Chewy, weirdly sweet, slightly crunchy"
    }
]

# Zugriff auf die Eigenschaft "spiciness" des zweiten Elements im Array
snacks[1]["spiciness"]
```

## Strukturierte vs. Unstrukturierte Daten

**Strukturierte Daten**  
- In einem festen Format organisiert, meist in Zeilen und Spalten (Tabellen, Datenbanken)  
- Leicht zu durchsuchen, zu analysieren und mit Standard-Tools zu verarbeiten  
- Beispiele: Excel-Sheets, SQL-Datenbanken, CSV-Dateien  

**Unstrukturierte Daten**  
- Keine vordefinierte Struktur oder festes Schema  
- Schwieriger zu verarbeiten; benötigen oft spezielle Tools oder Techniken  
- Beispiele: Text-Dokumente, Images, Videos, Audio-Dateien, Social-Media-Posts  

**Key Difference:**  
- **Strukturiert** = organisiert, klar aufgebaut  
- **Unstrukturiert** = flexibel, weniger klar strukturiert, aufwendiger zu analysieren
